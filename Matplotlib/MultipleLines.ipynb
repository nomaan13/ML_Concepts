```python
# Creating multiple lines in a single graph
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array([40,50,61,72,89,10])
y = np.array([10,25,15,60,50,20])

plt.xlabel("Age" , fontsize = 15)
plt.ylabel("Weight" ,fontsize = 20)
plt.title("MAPPING",fontsize = 30)


plt.plot(x ,c = 'orange' , label = 'MEN',linestyle='dotted')
plt.plot(y , c = 'Purple' , label = 'Female',linestyle = '--')


plt.legend()
plt.show()
```

```python
a = [19,20,38,41,5]
b = [12,22,56,43,31] 
c = [33,45,67,12,51]

plt.plot(a , c = 'red', linestyle  = 'dotted')
plt.plot(b,linestyle='--',c = 'Black')
plt.plot(c,c='orange',linestyle= '--')


plt.grid(c= 'green', linestyle = '--' )
plt.show()
```

```python
# IMport matplotlib

import matplotlib.pyplot as plt
```

```python
x = ["Apple","Mango","Grapes","PineApple","Orange",]
y = [120,40,60,85,20]

clr = ["red","yellow","green","brown","orange" ]

plt.xlabel('Fruits',fontsize = 15 )
plt.ylabel('Price',fontsize = 15)
plt.title('Fruits-Rate',fontsize = 20)


plt.bar(x,y,color=clr,label="Rate-List",align='edge',width=.5)
plt.legend()
plt.show()
```

```python
# imoport necessary libraries 

import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Data\logistic_regression_data.csv")

q = {"UK":0,"USA":1,"N":2}
df['Nationality'] = df['Nationality'].map(q)

p = {"NO":0,"YES":1}
df['Go'] = df['Go'].map(p)

# Differentiate the feature of the data


feature = ['Age', 'Experience' , 'Rank' ,'Nationality']
# label = ['Go']

X = df[feature]

y = df['Go']


dTree = DecisionTreeClassifier()
dTree = dTree.fit(X,y)

tree.plot_tree(dTree, feature_names=feature)

plt.show()
```

```python
df = pd.read_csv(r"D:\Data\logistic_regression_data.csv")

q = {"UK":0,"USA":1,"N":2}
df['Nationality'] = df['Nationality'].map(q)

p = {"NO":0,"YES":1}
df['Go'] = df['Go'].map(p)

# Differentiate the feature of the data


feature = ['Age', 'Experience' , 'Rank' ,'Nationality']
# label = ['Go']

X = df[feature]

y = df['Go']

dTree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)
dTree.fit(X, y)

# Plot Decision Tree
plt.figure(figsize=(12, 8))  # Set figure size
plot_tree(
    dTree, 
    feature_names=feature, 
    class_names=['No', 'Yes'], 
    filled=True, 
    rounded=True, 
    fontsize=15
)
plt.show()
```
